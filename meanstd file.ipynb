## standard deviation and mean of each columns

```python
dataT=data.T
j=0
meanlist=[]
stdlist=[]
meanstd=[]
for i in range(len(data.columns)):
    added=sum(data[i])
    len(data)
    mean=added/len(data)
    mean
    squaresum=sum(data[i]**2)
    std=squaresum/len(data)-mean**2
    meanstd.append(mean)
    meanstd.append(std)
    meanlist.append(mean)
    stdlist.append(std)
meanlist
stat=pd.DataFrame()
stat['mean']=meanlist
stat['std']=stdlist
stat
```
